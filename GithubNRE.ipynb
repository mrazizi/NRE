```
# visualize results

import json

resultFileName = "test_result/nyt_pcnn_att_pred.json"
testFileName = "data/nyt/test.json"

resultFile = open(resultFileName, 'r')
testResultJson = json.load(resultFile)

testFile = open(testFileName, 'r')
testJson = json.load(testFile)
```

```
testNumber = 2

currentHeadEntityID = testJson[testNumber].get('head').get('id')
currentTailEntityID = testJson[testNumber].get('tail').get('id')
currentEntityPair = currentHeadEntityID + "#" + currentTailEntityID

currentHeadEntityWord = testJson[testNumber].get('head').get('word')
currentTailEntityWord = testJson[testNumber].get('tail').get('word')

currentHeadEntityType = testJson[testNumber].get('head').get('type')
currentTailEntityType = testJson[testNumber].get('head').get('type')

currentSentence = testJson[testNumber].get('sentence')
currentRelation = testJson[testNumber].get('relation')

currentRelationNumber = testResultJson[testNumber].get('relation')
currentScore = testResultJson[testNumber].get('score')
# currentEntPairList = currentEntPair.split('#')
# currentHead = currentEntPairList[0]
# currentTail = currentEntPairList[1]

for item in testResultJson:
  if item.get('entpair') == currentEntityPair:
    print("head id: " + currentHeadEntityID)
    print("head word: " + currentHeadEntityWord)
    print("head type: " + currentHeadEntityType)
    
    print("\n")
    
    print("tail id : " + currentTailEntityID)
    print("tail word: " + currentTailEntityWord)
    print("tail type: " + currentTailEntityType)
    
    print("\n")
    
    print("sentence: " + currentSentence)
    print("relation: " + currentRelation)
    
    print("\n")
    
    print("relation number: " + str(currentRelationNumber))
    print("score: " + str(currentScore))
    
    
    print("\n---------------------------------------\n")
    
    
    
#   if item.get('head').get('id') == currentHead and item.get('tail').get('id') == currentTail:
#     print("head: " + item.get('head').get('word'))
#     print("tail: " + item.get('tail').get('word'))
#     print("relation: " + item.get('relation'))
#     print("sentence: " + item.get('sentence'))
#     print('\n')
```
